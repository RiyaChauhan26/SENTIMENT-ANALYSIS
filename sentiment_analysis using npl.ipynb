```python

# 📊 Sentiment Analysis using NLP (TextBlob)
# Dataset: Yelp Reviews


# Step 1: Import Libraries
import pandas as pd
import matplotlib.pyplot as plt
from textblob import TextBlob

# Step 2: Load Dataset
df = pd.read_csv("yelp.csv")

print("✅ Dataset Loaded Successfully\n")

# Show sample data
print("📌 Sample Data:\n")
print(df.head())

# Step 3: Check Columns
print("\n📌 Columns in Dataset:")
print(df.columns)

# Step 4: Define Sentiment Function
def get_sentiment(text):
    """
    This function returns sentiment of a given text
    """
    analysis = TextBlob(str(text))
    polarity = analysis.sentiment.polarity

    if polarity > 0:
        return "Positive"
    elif polarity < 0:
        return "Negative"
    else:
        return "Neutral"

# Step 5: Apply Sentiment Analysis
df["Sentiment"] = df["text"].apply(get_sentiment)

print("\n✅ Sentiment Analysis Applied Successfully\n")

# Show result
print("📌 Sample Output:\n")
print(df[["text", "Sentiment"]].head())

# Step 6: Count Sentiments
sentiment_counts = df["Sentiment"].value_counts()

print("\n📊 Sentiment Count:\n")
print(sentiment_counts)

# Step 7: Attractive Graph
plt.figure(figsize=(8,5))

# Bar plot
bars = plt.bar(sentiment_counts.index, sentiment_counts.values)

# Add values on top
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, int(yval),
             ha='center', va='bottom')

# Titles and labels
plt.title("📊 Sentiment Analysis of Yelp Reviews", fontsize=14)
plt.xlabel("Sentiment", fontsize=12)
plt.ylabel("Number of Reviews", fontsize=12)

# Grid for better look
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()

```

## Conclusion

The sentiment analysis was successfully performed on Yelp reviews.  
Most reviews were classified based on polarity score.  
This shows how NLP can help understand customer feedback easily.
